```python
texts = []
images=[]
labels =[]
language = []
ids = []
id = 0
with open(path,'r',encoding='utf-8') as f:
    f1_json = json.load(f)

    for line in tqdm(f1_json):
        text = line['text']
        label = line['label']

        #print(dest)

        if label == 0:
            dest = langauges[randrange(len(langauges))]
           # trans = model.translate(text,target_lang=dest)
            trans = translator.translate(text,src='en',dest=dest)

            labels.append(label)
            texts.append(trans)
            images.append(line['image_id'])
            language.append(dest)
            ids.append(id)
        else:
            for lan in langauges:
                dest = lan
                trans = translator.translate(text,src='en',dest=dest)
                labels.append(label)
                texts.append(trans)
                images.append(line['image_id'])
                language.append(dest)
                ids.append(id)

        id+=1


data = {'id':ids, 'text': texts,'iamge_id':images ,'label':labels,'language':language}
df = pd.DataFrame(data = data)

```

```python
def edit_text(text):
    return text.text

# Apply the function to the 'text' column
df['text'] = df['text'].apply(edit_text)
```
